## 测试时间

```python
from cnswd.websource.wy import fetch_ohlcv
```

```python
# 注意不要使用期间有分红派息的股票
df = fetch_ohlcv('600645','2019-01-01','2019-06-30')
```

```python
# 确保按升序排列

df.sort_index(inplace=True)
```

```python
df.tail()
```

```python
ohlc = df.loc[:,['开盘价','最高价','最低价','收盘价']]
```

```python
ohlc.columns = ['open','high','low','close']
```

### 自然序号

```python
ohlc.reset_index(drop=True, inplace=True)
```

```python
ohlc.tail()
```

```python
qf=cf.QuantFig(ohlc,title='量化分析图',legend='top',name='中源协和')
```

```python
qf.add_bollinger_bands()
```

```python
qf.iplot()
```

```python
qf.figure()
```
