```python
from zipline.pipeline.data import CNEquityPricing
from zipline.pipeline import Pipeline
from zipline.research import run_pipeline, select_output_by, symbol
```

```python
def make_pipeline():
    return Pipeline(columns={
        'open': CNEquityPricing.open.latest,
        'high': CNEquityPricing.high.latest,
        'low': CNEquityPricing.low.latest,
        'close': CNEquityPricing.close.latest,
        'volume': CNEquityPricing.volume.latest,
    })
```

```python
df = run_pipeline(make_pipeline(), '2020-05-26', '2020-07-03')
```

```python
# 平安银行 2020-05-28 每股分红 0.218
# TODO:需要检验调整是否正确
select_output_by(df, assets=symbol('000001'))
```
