```python
import pandas as pd
```

```python
speed = [0.1, 17.5, 40, 48, 52, 69, 88]
lifespan = [2, 8, 70, 1.5, 25, 12, 28]
index = ["蜗牛","猪","大象", "兔子","长颈鹿","土狼","马"]
df = pd.DataFrame({'速度': speed,
                   '寿命': lifespan}, index=index)
ax = df.plot.bar(rot=0)
```

```python
axes = df.plot.bar(rot=0, subplots=True)
axes[1].legend(loc=2)
```
